```python
all_contacts_file = 'all-contacts_20191118.csv'
```

```python
after_underscore = all_contacts_file.find('_')+1
before_ext = all_contacts_file.find('.csv')
contacts_file_date = all_contacts_file[after_underscore:before_ext]
contacts_file_date
```

```python
df = pd.read_csv(all_contacts_file)
```

```python
df = df['Email']
```

```python
email_array = df.to_numpy()
```

```python
email_endings = []
for i in tqdm(range(len(email_array))):
    email = str(email_array[i]).lower().strip()
    where_at = email.find('@')
    ending = email[where_at+1:]
    email_endings.extend([ending])
unique_emails = list(dict.fromkeys(email_endings))
```

```python
len(unique_emails)
```

```python
econame_file = 'EmailCompanyName_{}.json'.format(contacts_file_date)
if os.path.exists(econame_file):
    print('Dictionary file already exists.')
    with open(econame_file,'r+') as ecf:
        ecd = json.load(ecf)
        for i in tqdm(range(len(unique_emails))):
        #for i in range(2):
            domain_info = hunter.domain_search(unique_emails[i])
            org = domain_info.get('organization')
            webmail = domain_info.get('webmail')
            if unique_emails[i] not in ecd and webmail == False and org != None:
                ecd[unique_emails[i]] = org
        ecf.seek(0)
        json.dump(ecd,ecf)
        ecf.truncate()
else:
    print('Dictionary file does not exist, but has been created.')
    with open(econame_file,'w') as ecf:
        ecd = dict()
        for i in tqdm(range(len(unique_emails))):
        #for i in range(2):
            domain_info = hunter.domain_search(unique_emails[i])
            org = domain_info.get('organization')
            webmail = domain_info.get('webmail')
            if unique_emails[i] not in ecd and webmail == False and org != None:
                ecd[unique_emails[i]] = org
        json.dump(ecd,ecf)
```
