```python
import pandas as pd
import numpy as np
from tqdm import tnrange, tqdm_notebook as tqdm
import datetime
```

```python
df = pd.read_csv('all-contacts_20190716.csv')
```

```python
df.head(0)
```

```python
cols = ['Email','Industry']
```

```python
df = df[cols]
df.head(0)
```

```python
array = df.to_numpy()
```

```python
array[170, 1] == 'Shipper/Retailer/Manufacturer/BCO'
```

```python
df.shape
```

```python
start = datetime.datetime.now()
print('Start time: {}'.format(start))
shippers_array = []

for i in tqdm(range(df.shape[0])):
    ind = str(array[i][1])
    if ind == 'Shipper/Retailer/Manufacturer/BCO':
        shippers_array.extend([array[i]])

end = datetime.datetime.now()
print('Elapsed time: {}'.format(end - start))
```

```python
len(shippers_array)
```

```python
email_endings = []
for i in tqdm(range(len(shippers_array))):
    email = str(shippers_array[i][0]).lower().strip()
    where_at = email.find('@')
    ending = email[where_at+1:]
    email_endings.extend([ending])
```

```python
unique_emails = list(dict.fromkeys(email_endings))
len(unique_emails)
```

```python
d = {'':''}
for i in tqdm(range(len(unique_emails))):
    count = 0
    domain = unique_emails[i]
    for j in range(len(email_endings)):
        if email_endings[j] == domain:
            count += 1
    d[domain] = count
d.pop('')
```

```python
d
```

```python
#e_df = pd.DataFrame(d.items())
#e_df.sort_values[by=, ascending=False, inplace=True]
s = pd.Series(d, name='Count')
s.index.name = 'Domain'
s.reset_index()
s = pd.DataFrame(s, index=None)
s.sort_values(by=['Count'], ascending=False, inplace=True)
```

```python
s.to_csv('srmb_emails_20190716.csv')
```
